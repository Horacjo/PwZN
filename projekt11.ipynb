```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
with open('facebook_combined.txt') as f:
    lines = f.readlines()

myList = [line.strip().split() for line in lines]

g = nx.Graph()
g.add_edges_from(myList)
```

```python
print(f'Liczba wierzchołków: {g.number_of_nodes()}')
print(f'Liczba krawędzi: {g.number_of_edges()}')
print(f'Spójny: {nx.is_connected(g)}')
print(f'Liczba spójnych komponentów: {nx.number_connected_components(g)}')
print(f'Średnia długość najkrótszej ścieżki: {nx.average_shortest_path_length(g, weight=None, method=None)}')
```

```python
###Rozkład stopni wierzchołków
degree_freq = nx.degree_histogram(g)
degrees = range(len(degree_freq))
plt.loglog(degrees[3:], degree_freq[3:],'go-') 
plt.xlabel('k')
plt.ylabel('P(k)')
plt.title('Degree Distribution')
```

```python
###Wizualizacjagrafu
plt.rcParams["figure.figsize"] = [20, 10]
nx.draw(g,pos=nx.spring_layout(g), node_size = 5, width = 0.1, alpha = 0.4)
```
